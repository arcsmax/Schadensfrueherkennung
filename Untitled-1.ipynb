```python
# Code Zelle 2: Daten laden (Überarbeitet für PHM 2023 Struktur - JETZT MIT KORREKTUR FÜR ORDNERNAMEN)
import os
import re # Für das Parsen der Dateinamen
import pandas as pd # Sicherstellen, dass Pandas hier verfügbar ist
import numpy as np  # Sicherstellen, dass NumPy hier verfügbar ist

# Konstanten aus der Recherche (falls nicht schon global definiert)
SAMPLING_RATE_HZ = 20480
GEAR_RATIO_INPUT_TO_OUTPUT = 1.8

def parse_filename(filename: str) -> dict:
    """
    Extrahiert Betriebsbedingungen und Wiederholungsnummer aus dem Dateinamen.
    Beispiel: V100_200N_2.txt -> {'rpm_input': 100, 'torque_output_nm': 200, 'repetition': 2}
    """
    # Das Muster berücksichtigt jetzt V oder v am Anfang und .txt oder .TXT am Ende
    match = re.match(r'V(\d+)_(\d+)N_(\d+)\.txt', filename, re.IGNORECASE)
    if match:
        return {
            'rpm_input': int(match.group(1)),
            'torque_output_nm': int(match.group(2)),
            'repetition': int(match.group(3))
        }
    print(f"WARNUNG: Dateiname {filename} konnte nicht mit dem Muster V<RPM>_<TORQUE>N_<REP>.txt geparst werden.")
    return {}

def parse_folder_name(folder_name: str) -> int:
    """
    Extrahiert den numerischen Schädigungsgrad aus dem Ordnernamen.
    Beispiele:
    'Pitting_degradation_level_2' -> 2
    'Pitting_degradation_level_0 (Healthy)' -> 0
    'level_5' -> 5 (falls ein anderes Projekt dieses Format verwendet)
    Gibt -1 zurück, falls kein Level extrahiert werden kann.
    """
    # Versuche, eine Zahl am Ende von "_level_X" zu finden
    match = re.search(r'_level_(\d+)', folder_name, re.IGNORECASE)
    if match:
        return int(match.group(1))
    
    # Fallback für Ordner wie "Healthy" oder "level_0" direkt, falls das obige nicht passt
    # Dieser Teil ist jetzt spezifischer durch das obige Regex abgedeckt.
    # Man könnte hier noch expliziter nach "Healthy" suchen, um es Level 0 zuzuordnen,
    # falls das "_level_0" nicht immer vorhanden ist.
    if "healthy" in folder_name.lower() and "_level_0" in folder_name.lower(): # Speziell für "Pitting_degradation_level_0 (Healthy)"
        return 0
    if "healthy" in folder_name.lower() and not re.search(r'_level_', folder_name, re.IGNORECASE): # Falls es nur "Healthy" hieße
         return 0

    print(f"WARNUNG: Schädigungsgrad konnte nicht aus Ordnernamen '{folder_name}' extrahiert werden.")
    return -1 # Signalisiert einen Fehler oder einen nicht erkannten Ordner

def load_single_phm_file(filepath: str, filename_info: dict) -> pd.DataFrame:
    """
    Lädt eine einzelne .txt-Datei und fügt Metadaten hinzu.
    Die .txt-Dateien enthalten Leerzeichen-separierte Werte.
    """
    try:
        # Die Annahme ist, dass die Trennzeichen variable Leerzeichen sind, daher \s+
        df = pd.read_csv(filepath, sep=r'\s+', header=None,
                         names=['acc_x', 'acc_y', 'acc_z', 'tacho'],
                         engine='python') # 'python' engine ist manchmal robuster bei komplexen Separatoren
        
        # Überprüfen, ob die erwartete Anzahl von Spalten geladen wurde
        if df.shape[1] != 4:
            print(f"WARNUNG: Unerwartete Spaltenanzahl ({df.shape[1]}) in Datei {filepath}. Erwartet 4.")
            # Man könnte hier versuchen, die Zeilen zu filtern, die nicht 4 Spalten haben,
            # oder die Datei als fehlerhaft markieren.
            # df = df.dropna(thresh=4) # Zeilen verwerfen, die nicht mindestens 4 Werte haben (nicht ideal)
            # return pd.DataFrame() # Datei überspringen

        df['rpm_input'] = filename_info.get('rpm_input')
        df['torque_output_nm'] = filename_info.get('torque_output_nm')
        df['repetition'] = filename_info.get('repetition')
        df['filename'] = os.path.basename(filepath)

        num_samples = len(df)
        df['timestamp'] = np.arange(num_samples) / SAMPLING_RATE_HZ

        return df
    except pd.errors.EmptyDataError:
        print(f"WARNUNG: Datei {filepath} ist leer oder enthält nur Header.")
        return pd.DataFrame()
    except Exception as e:
        print(f"Fehler beim Laden oder Verarbeiten von {filepath}: {e}")
        return pd.DataFrame()

def load_phm_data_challenge_2023(dataset_base_path: str) -> pd.DataFrame:
    """
    Lädt alle relevanten Daten des PHM North America 2023 Data Challenge.
    Passt sich an Ordnernamen wie 'Pitting_degradation_level_X' an.
    """
    all_data = []
    print(f"Durchsuche Basisverzeichnis: {dataset_base_path}")

    if not os.path.isdir(dataset_base_path):
        print(f"FEHLER: Das Basisverzeichnis '{dataset_base_path}' existiert nicht oder ist keine Directory.")
        return pd.DataFrame()

    for folder_name in os.listdir(dataset_base_path):
        folder_path = os.path.join(dataset_base_path, folder_name)
        if os.path.isdir(folder_path):
            pitting_level = parse_folder_name(folder_name) # Verwendet die neue Parsing-Funktion
            
            if pitting_level != -1: # Erfolgreich geparst
                print(f"Lade Daten für Schädigungsgrad: {pitting_level} aus Ordner: {folder_name}")
                
                files_in_folder = os.listdir(folder_path)
                if not files_in_folder:
                    print(f"WARNUNG: Ordner {folder_name} ist leer.")
                    continue

                for filename in files_in_folder:
                    if filename.lower().endswith('.txt'):
                        filepath = os.path.join(folder_path, filename)
                        filename_info = parse_filename(filename)
                        if filename_info: # Nur wenn Dateiname erfolgreich geparst wurde
                            df_file = load_single_phm_file(filepath, filename_info)
                            if not df_file.empty:
                                df_file['pitting_level'] = pitting_level
                                all_data.append(df_file)
            else:
                # Diese Nachricht wird jetzt von parse_folder_name übernommen, wenn das Parsen fehlschlägt.
                # Man könnte hier zusätzlich loggen oder bestimmte Ordner explizit ignorieren (z.B. '.DS_Store')
                if not folder_name.startswith('.'): # Ignoriere versteckte Systemordner
                     print(f"INFO: Ordner '{folder_name}' wird übersprungen (Schädigungsgrad nicht erkannt).")
        else:
            if not folder_name.startswith('.'): # Ignoriere versteckte Systemdateien
                print(f"INFO: Eintrag '{folder_name}' im Basisverzeichnis ist kein Ordner und wird übersprungen.")


    if not all_data:
        print("WARNUNG: Keine Daten geladen. Überprüfen Sie den dataset_base_path und die Ordnerstruktur.")
        # Hier KEINE Dummy-Daten mehr erstellen, da wir echte Daten laden wollen.
        # Der Fehler sollte jetzt klarer sein.
        return pd.DataFrame()

    combined_df = pd.concat(all_data, ignore_index=True)

    print(f"\nGesamte Datenstruktur nach dem Laden ({len(all_data)} Dateien erfolgreich verarbeitet):")
    if not combined_df.empty:
        print(combined_df.head())
        print(f"\nForm der Daten: {combined_df.shape}")
        print(f"\nDatentypen:\n{combined_df.dtypes}")
        # Überprüfe auf NaN-Werte, die durch fehlerhaftes Laden entstehen könnten
        for col in ['acc_x', 'acc_y', 'acc_z', 'tacho']:
            if combined_df[col].isnull().any():
                print(f"WARNUNG: Spalte {col} enthält NaN-Werte nach dem Laden. Überprüfen Sie die .txt-Dateien und die Ladefunktion `load_single_phm_file`.")

        if 'pitting_level' in combined_df.columns:
            print(f"\nVerteilung der Schädigungsgrade:\n{combined_df['pitting_level'].value_counts().sort_index()}")
        if 'rpm_input' in combined_df.columns:
            print(f"\nEinzigartige Eingangsdrehzahlen (RPM):\n{sorted(combined_df['rpm_input'].unique())}")
        if 'torque_output_nm' in combined_df.columns:
            print(f"\nEinzigartige Ausgangsdrehmomente (Nm):\n{sorted(combined_df['torque_output_nm'].unique())}")
    else:
        print("INFO: Der kombinierte DataFrame ist leer, obwohl Dateien verarbeitet werden sollten. Überprüfen Sie die Warnungen oben.")
        
    return combined_df
# Daten laden und überprüfen
DATASET_PATH_PHM2023_NA = "/Users/maxbook/Documents/Data_Challenge_PHM2023_training_data"
raw_data_df = load_phm_data_challenge_2023(DATASET_PATH_PHM2023_NA)

# Überprüfen, ob Daten geladen wurden, bevor fortgefahren wird
if raw_data_df.empty:
    print("---------------------------------------------------------------------------")
    print("FEHLER: KEINE DATEN GELADEN. DAS NOTEBOOK KANN NICHT SINNVOLL FORTGESETZT WERDEN.")
    print("Bitte überprüfen Sie:")
    print(f"1. Ob der Pfad '{DATASET_PATH_PHM2023_NA}' korrekt ist und existiert.")
    print("2. Ob die Ordnerstruktur darunter den Schädigungsgrad-Ordnern entspricht (z.B. 'Pitting_degradation_level_0 (Healthy)', 'Pitting_degradation_level_1', etc.).")
    print("3. Ob diese Ordner .txt-Dateien mit den Vibrationsdaten enthalten.")
    print("4. Die Konsolenausgaben oben auf 'WARNUNG' oder 'FEHLER' Meldungen beim Laden.")
    print("---------------------------------------------------------------------------")
    # raise ValueError("Keine Daten geladen. Bitte überprüfen Sie den Pfad und die Ladefunktion.") # Auskommentiert, um das Notebook nicht sofort zu stoppen
else:
    print("\nDaten erfolgreich geladen und DataFrame `raw_data_df` erstellt.")
```

```python
# Code Zelle: Charakteristische Frequenzen berechnen und anwenden
# Diese Zelle wird ausgeführt, nachdem raw_data_df_loaded erfolgreich aus der Parquet-Datei geladen wurde.

import numpy as np
import pandas as pd # Sicherstellen, dass Pandas importiert ist

# Annahme: raw_data_df_loaded wurde in der vorherigen Zelle erfolgreich geladen und ist verfügbar.
# Wenn du mit 'raw_data_df' weiterarbeiten möchtest, kopiere oder benenne es um:
if 'raw_data_df_loaded' in locals() and not raw_data_df_loaded.empty:
    raw_data_df = raw_data_df_loaded.copy() # Arbeite mit einer Kopie oder direkt mit raw_data_df_loaded
    print("DataFrame `raw_data_df_loaded` wurde nach `raw_data_df` kopiert für weitere Verwendung.")
elif 'raw_data_df' not in locals() or raw_data_df.empty:
    print("FEHLER: Weder `raw_data_df_loaded` noch `raw_data_df` sind gültige DataFrames. "
          "Bitte stelle sicher, dass die Daten korrekt geladen wurden.")
    # Hier könnte man das Notebook stoppen oder einen Dummy-DataFrame für Testzwecke erstellen
    # Für dieses Beispiel wird angenommen, dass raw_data_df jetzt existiert.
    # raise ValueError("Kein gültiger DataFrame zum Arbeiten vorhanden.")

# --- Definition der Getriebeparameter und Berechnungsfunktion ---

# Getriebeparameter (konstant für diesen Datensatz, basierend auf deiner Recherche)
Z_INPUT_GEAR = 40  # Zähnezahl antreibendes Zahnrad (Pinion)
Z_OUTPUT_GEAR = 72 # Zähnezahl angetriebenes Zahnrad (Gear)

# Übersetzungsverhältnis (Input:Output)
# Diese Konstante sollte idealerweise einmalig nach dem Laden der Daten oder global definiert werden.
GEAR_RATIO_INPUT_TO_OUTPUT = 1.8 

# Konsistenzprüfung des gegebenen Übersetzungsverhältnisses mit den Zähnezahlen
# (Diese Prüfung kann einmalig nach Definition der Konstanten erfolgen)
calculated_ratio_check = Z_OUTPUT_GEAR / Z_INPUT_GEAR
if abs(calculated_ratio_check - GEAR_RATIO_INPUT_TO_OUTPUT) > 1e-3:
    print(f"WARNUNG (Globale Prüfung): Das gegebene Übersetzungsverhältnis ({GEAR_RATIO_INPUT_TO_OUTPUT}) "
          f"weicht vom berechneten Verhältnis ({calculated_ratio_check:.3f}) basierend auf den Zähnezahlen ab.")

def calculate_gear_characteristic_frequencies(
    input_shaft_speed_rpm: float,
    z_input: int = Z_INPUT_GEAR,
    z_output: int = Z_OUTPUT_GEAR,
    gear_ratio: float = GEAR_RATIO_INPUT_TO_OUTPUT # Verwende die globale Konstante
) -> dict:
    """
    Berechnet charakteristische Getriebefrequenzen (Wellen und Zahneingriff).

    Args:
        input_shaft_speed_rpm (float): Drehzahl der Eingangswelle in RPM.
        z_input (int): Zähnezahl des antreibenden Zahnrads.
        z_output (int): Zähnezahl des angetriebenen Zahnrads.
        gear_ratio (float): Übersetzungsverhältnis (Eingang zu Ausgang).

    Returns:
        dict: Ein Dictionary mit den berechneten Frequenzen in Hz:
              'input_shaft_hz', 'output_shaft_hz', 'gmf_hz'.
              Gibt Frequenzen als 0.0 zurück, wenn input_shaft_speed_rpm <= 0.
    """
    if input_shaft_speed_rpm <= 0:
        return {
            'input_shaft_hz': 0.0,
            'output_shaft_hz': 0.0,
            'gmf_hz': 0.0
        }

    frequencies = {}
    input_shaft_speed_hz = input_shaft_speed_rpm / 60.0
    
    if gear_ratio == 0: # Vermeide Division durch Null
        output_shaft_speed_hz = 0.0
        print(f"WARNUNG: Übersetzungsverhältnis ist 0 bei {input_shaft_speed_rpm} RPM. Output Shaft Hz auf 0 gesetzt.")
    else:
        output_shaft_speed_hz = input_shaft_speed_hz / gear_ratio

    frequencies['input_shaft_hz'] = round(input_shaft_speed_hz, 3)
    frequencies['output_shaft_hz'] = round(output_shaft_speed_hz, 3)
    
    gmf_from_input = input_shaft_speed_hz * z_input
    frequencies['gmf_hz'] = round(gmf_from_input, 3)
    
    # Erweiterter Konsistenzcheck für GMF
    if z_output > 0 and output_shaft_speed_hz > 0: # Nur prüfen, wenn Werte sinnvoll sind
        gmf_from_output = output_shaft_speed_hz * z_output
        # Relative Toleranz, um bei hohen Frequenzen nicht zu empfindlich zu sein
        if abs(gmf_from_input - gmf_from_output) > (0.001 * gmf_from_input + 1e-3) : # 0.1% relative + kleine absolute Toleranz
            print(f"HINWEIS (GMF-Prüfung bei {input_shaft_speed_rpm:.0f} RPM): "
                  f"GMF (Input-basiert)={gmf_from_input:.2f} Hz, "
                  f"GMF (Output-basiert)={gmf_from_output:.2f} Hz. "
                  f"Kleine Abweichungen können durch Rundung des Übersetzungsverhältnisses entstehen.")
            
    # Hinweis zu Lagerfrequenzen (da keine genauen Daten vorhanden)
    # frequencies['bearing_info'] = "Spezifische Lagerdaten nicht verfügbar für exakte Berechnung."
            
    return frequencies

# --- Beispielhafte Anwendung für eine mittlere Drehzahl aus den Daten ---
char_freqs_example_output = {} # Initialisiere für den Fall, dass keine Daten/RPMs gefunden werden

if 'raw_data_df' in locals() and not raw_data_df.empty and 'rpm_input' in raw_data_df.columns:
    unique_rpms_list = sorted(raw_data_df['rpm_input'].unique())
    
    if unique_rpms_list:
        # Wähle eine mittlere, positive Drehzahl für das Beispiel
        positive_rpms = [rpm for rpm in unique_rpms_list if rpm > 0]
        if positive_rpms:
            example_rpm_value = positive_rpms[len(positive_rpms) // 2]
            char_freqs_example_output = calculate_gear_characteristic_frequencies(example_rpm_value)
            
            print(f"\nBeispielhafte charakteristische Frequenzen bei {example_rpm_value:.0f} RPM Eingangswelle:")
            for name, val in char_freqs_example_output.items():
                print(f"  {name}: {val} Hz")
        else:
            print("WARNUNG: Keine positiven Drehzahlen im DataFrame für Beispielberechnung gefunden.")
    else:
        print("WARNUNG: Keine einzigartigen Drehzahlen im DataFrame für Beispielberechnung gefunden.")
else:
    print("WARNUNG: DataFrame 'raw_data_df' oder Spalte 'rpm_input' nicht für Beispielberechnung verfügbar.")

# --- (Optional aber empfohlen) Charakteristische Frequenzen für jede Messung vorab berechnen ---
# Dies erstellt ein Dictionary, das jeder einzigartigen Kombination von 
# 'filename' (als eindeutiger Identifier für eine Messung) die zugehörigen Frequenzen zuordnet.
# Das ist effizienter, als die Frequenzen jedes Mal neu zu berechnen, wenn sie für einen Plot benötigt werden.

calculated_frequencies_per_file = {}

if 'raw_data_df' in locals() and not raw_data_df.empty:
    # Wir benötigen nur eine Zeile pro Datei, um die RPM für diese Datei zu bekommen
    # anstatt über alle ~260 Mio. Zeilen zu iterieren.
    # `drop_duplicates` behält das erste Vorkommen.
    # Wichtig: 'filename' und 'rpm_input' müssen vorhanden sein.
    if 'filename' in raw_data_df.columns and 'rpm_input' in raw_data_df.columns:
        # Erstelle einen kleineren DataFrame mit nur den einzigartigen Kombinationen von filename und rpm_input
        # Dies vermeidet die Iteration über Millionen von Zeilen, wenn nur wenige hundert Dateien existieren.
        unique_files_rpm_df = raw_data_df[['filename', 'rpm_input']].drop_duplicates().reset_index(drop=True)
        print(f"\nBerechne charakteristische Frequenzen für {len(unique_files_rpm_df)} einzigartige Messdateien...")

        for index, row in unique_files_rpm_df.iterrows():
            filename = row['filename']
            rpm = row['rpm_input']
            if pd.notna(rpm) and pd.notna(filename): # Stelle sicher, dass Werte nicht NaN sind
                 calculated_frequencies_per_file[filename] = calculate_gear_characteristic_frequencies(rpm)
            else:
                print(f"WARNUNG: Überspringe Datei '{filename}' oder RPM '{rpm}' aufgrund von NaN-Werten.")


        print(f"Berechnung abgeschlossen. {len(calculated_frequencies_per_file)} Frequenz-Sets gespeichert.")
        
        # Beispielhafte Ausgabe für einige wenige Dateien:
        if calculated_frequencies_per_file:
            print("\nBeispiel-Frequenzen für einige Dateien:")
            count = 0
            for filename, freqs in calculated_frequencies_per_file.items():
                print(f"  Datei: {filename} (RPM: {unique_files_rpm_df[unique_files_rpm_df['filename'] == filename]['rpm_input'].iloc[0]}) -> Frequenzen: {freqs}")
                count += 1
                if count >= 3: # Zeige nur die ersten paar
                    break
        else:
            print("Keine Frequenzen berechnet (möglicherweise keine gültigen RPM-Werte).")
            
    else:
        print("WARNUNG: Spalten 'filename' oder 'rpm_input' nicht im DataFrame vorhanden. "
              "Vorabberechnung der Frequenzen pro Datei übersprungen.")
else:
    print("WARNUNG: DataFrame 'raw_data_df' nicht verfügbar für die Vorabberechnung der Frequenzen.")
```

```python
# Code Zelle 4, 5, 6: Verbesserte Visualisierung von Rohsignalen, FFTs, Spektrogrammen und Tacho

# Importiere notwendige Bibliotheken, falls nicht schon geschehen
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.fft import fft, fftfreq
import scipy.signal as signal # Für STFT

# Annahmen:
# - raw_data_df: Dein geladener und verfügbarer DataFrame
# - calculated_frequencies_per_file: Das Dictionary mit den vorab berechneten Frequenzen pro Datei
# - SAMPLING_RATE_HZ, NYQUIST_FREQ_HZ, GEAR_RATIO_INPUT_TO_OUTPUT: global definierte Konstanten


# --- Hilfsfunktion zur Auswahl repräsentativer Dateien ---
def get_representative_file_info(
    df: pd.DataFrame,
    pitting_level_target: int,
    rpm_target: int = None, # Optional: spezifische RPM
    torque_target: int = None # Optional: spezifisches Drehmoment
) -> pd.Series:
    """
    Sucht eine repräsentative Datei (als Series) basierend auf Pitting Level und optional RPM/Torque.
    Gibt die erste gefundene Übereinstimmung zurück oder None.
    """
    if df.empty:
        return None

    conditions = (df['pitting_level'] == pitting_level_target)
    
    # Finde verfügbare RPMs und Torques für den Ziel-Pittinglevel
    available_rpms_for_level = sorted(df[conditions]['rpm_input'].unique())
    available_torques_for_level = sorted(df[conditions]['torque_output_nm'].unique())

    if not available_rpms_for_level or not available_torques_for_level:
        print(f"WARNUNG: Keine Daten für Pitting Level {pitting_level_target} gefunden.")
        return None

    if rpm_target is None or rpm_target not in available_rpms_for_level:
        # Wähle eine mittlere RPM, falls kein Ziel oder Ziel nicht verfügbar
        rpm_to_use = available_rpms_for_level[len(available_rpms_for_level) // 2]
    else:
        rpm_to_use = rpm_target
    
    conditions &= (df['rpm_input'] == rpm_to_use)
    
    # Aktualisiere verfügbare Torques basierend auf gewählter RPM
    available_torques_for_rpm_and_level = sorted(df[conditions]['torque_output_nm'].unique())
    if not available_torques_for_rpm_and_level:
        print(f"WARNUNG: Keine Daten für Pitting Level {pitting_level_target} bei RPM {rpm_to_use} gefunden.")
        return None

    if torque_target is None or torque_target not in available_torques_for_rpm_and_level:
        # Wähle ein mittleres Drehmoment
        torque_to_use = available_torques_for_rpm_and_level[len(available_torques_for_rpm_and_level) // 2]
    else:
        torque_to_use = torque_target
        
    conditions &= (df['torque_output_nm'] == torque_to_use)

    representative_sample = df[conditions].head(1) # Nimm die erste Datei, die alle Kriterien erfüllt
    
    if representative_sample.empty:
        print(f"WARNUNG: Keine Datei für Pitting Lvl {pitting_level_target}, RPM {rpm_to_use}, Torque {torque_to_use} gefunden.")
        return None
    
    return representative_sample.iloc[0] # Gibt die Zeile als Series zurück


# --- Verbesserte Plot-Funktionen ---

def plot_time_signals(
    data_segment_df: pd.DataFrame,
    sensor_cols: list,
    title_prefix: str,
    y_axis_label: str = "Amplitude (g?)" # TODO: Einheit klären und anpassen!
):
    """
    Plottet die Zeitreihen für die gegebenen Sensoren aus einem DataFrame-Segment.
    """
    if data_segment_df.empty:
        print(f"Keine Daten zum Plotten für '{title_prefix}'.")
        return

    plt.figure(figsize=(16, len(sensor_cols) * 2.5))
    for i, sensor_col in enumerate(sensor_cols):
        if sensor_col not in data_segment_df.columns:
            print(f"Sensor '{sensor_col}' nicht in den Daten für '{title_prefix}'.")
            continue
        plt.subplot(len(sensor_cols), 1, i + 1)
        plt.plot(data_segment_df['timestamp'], data_segment_df[sensor_col], linewidth=0.8)
        plt.title(f"{title_prefix} - Sensor: {sensor_col}", fontsize=10)
        plt.xlabel("Zeit (s)", fontsize=9)
        plt.ylabel(y_axis_label, fontsize=9)
        plt.xticks(fontsize=8)
        plt.yticks(fontsize=8)
        plt.grid(True, linestyle=':', alpha=0.7)
    plt.tight_layout(rect=[0, 0, 1, 0.97]) # Platz für Haupttitel lassen
    plt.suptitle(title_prefix, fontsize=12, fontweight='bold')
    plt.show()


def plot_fft_signals(
    data_segment_df: pd.DataFrame,
    sensor_col: str,
    sampling_rate: float,
    char_freqs: dict,
    title_info: str,
    n_fft_points: int = 8192, # Mehr Punkte für bessere Frequenzauflösung
    max_freq_display: float = 2000,
    use_log_scale_y: bool = False,
    y_axis_label: str = "Amplitude"
):
    """
    Plottet das Frequenzspektrum (FFT) für einen gegebenen Sensor aus einem DataFrame-Segment.
    """
    if data_segment_df.empty or sensor_col not in data_segment_df.columns:
        print(f"Keine Daten oder Sensor '{sensor_col}' nicht vorhanden für FFT-Plot: {title_info}.")
        return

    signal_data = data_segment_df[sensor_col].head(n_fft_points).values
    if len(signal_data) < n_fft_points:
        print(f"Nicht genügend Samples ({len(signal_data)}) für FFT für {title_info}. Benötigt {n_fft_points}.")
        return

    yf = fft(signal_data)
    xf = fftfreq(n_fft_points, 1 / sampling_rate)[:n_fft_points // 2]
    magnitude = 2.0 / n_fft_points * np.abs(yf[0:n_fft_points // 2])

    plt.figure(figsize=(14, 6))
    plt.plot(xf, magnitude, linewidth=0.9)
    plt.title(f"FFT - Sensor: {sensor_col} ({title_info})", fontsize=10)
    plt.xlabel("Frequenz (Hz)", fontsize=9)
    plt.ylabel(y_axis_label, fontsize=9)
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)
    plt.grid(True, linestyle=':', alpha=0.7)

    added_labels = set() # Um doppelte Labels in der Legende zu vermeiden
    if char_freqs:
        for freq_name, freq_val in char_freqs.items():
            if 0 < freq_val <= xf.max():
                label = f"{freq_name}: {freq_val:.1f} Hz"
                if label not in added_labels:
                    plt.axvline(x=freq_val, color='r', linestyle='--', alpha=0.6, label=label, linewidth=1)
                    added_labels.add(label)
                else: # Nur Linie zeichnen ohne Label, wenn schon vorhanden
                    plt.axvline(x=freq_val, color='r', linestyle='--', alpha=0.6, linewidth=1)

    if added_labels: # Legende nur anzeigen, wenn Frequenzen geplottet wurden
        plt.legend(fontsize='x-small', loc='upper right')

    plt.xlim(0, min(max_freq_display, xf.max()))
    if use_log_scale_y:
        plt.yscale('log')
        # Mindestwert für log y-Achse setzen, um Fehler bei Nullen zu vermeiden
        min_val_for_log = np.min(magnitude[magnitude > 0]) / 10 if np.any(magnitude > 0) else 1e-6
        plt.ylim(bottom=min_val_for_log)
        
    plt.show()


def plot_stft_signal(
    data_segment_df: pd.DataFrame,
    sensor_col: str,
    sampling_rate: float,
    title_info: str,
    n_stft_samples: int = None, # Standardmäßig 1-2 Sekunden
    nperseg: int = 512,
    noverlap_ratio: float = 0.75, # Prozentualer Überlapp
    max_freq_display: float = 2000
):
    """Plottet das Spektrogramm (STFT) für einen gegebenen Sensor."""
    if data_segment_df.empty or sensor_col not in data_segment_df.columns:
        print(f"Keine Daten oder Sensor '{sensor_col}' nicht vorhanden für STFT-Plot: {title_info}.")
        return
        
    if n_stft_samples is None:
        n_stft_samples = int(sampling_rate * 1.5) # Standardmäßig 1.5 Sekunden Daten

    signal_data = data_segment_df[sensor_col].head(n_stft_samples).values
    
    min_len_for_stft = nperseg # Mindestlänge für STFT ist die Fensterlänge
    if len(signal_data) < min_len_for_stft:
        print(f"Nicht genügend Samples ({len(signal_data)}) für STFT für {title_info}. Benötigt mind. {min_len_for_stft}.")
        return

    noverlap_val = int(nperseg * noverlap_ratio)
    f_stft, t_stft, Zxx_stft = signal.stft(signal_data, fs=sampling_rate, nperseg=nperseg, noverlap=noverlap_val)
    
    plt.figure(figsize=(14, 6))
    actual_max_freq_display = min(max_freq_display, f_stft.max())
    relevant_f_indices_stft = np.where(f_stft <= actual_max_freq_display)[0]

    if len(relevant_f_indices_stft) > 0:
        # Begrenze vmax, um die Farbskala aussagekräftiger zu machen (Ausreißer ignorieren)
        vmax_val = np.percentile(np.abs(Zxx_stft[relevant_f_indices_stft, :]), 99.5) if Zxx_stft.size > 0 else None
        
        plt.pcolormesh(t_stft, f_stft[relevant_f_indices_stft],
                       np.abs(Zxx_stft[relevant_f_indices_stft, :]),
                       shading='gouraud', cmap='viridis', vmax=vmax_val)
        plt.title(f"Spektrogramm - Sensor: {sensor_col} ({title_info})", fontsize=10)
        plt.ylabel('Frequenz (Hz)', fontsize=9)
        plt.xlabel('Zeit (s)', fontsize=9)
        plt.xticks(fontsize=8)
        plt.yticks(fontsize=8)
        plt.colorbar(label='Intensität')
        plt.ylim(0, actual_max_freq_display)
        plt.show()
    else:
        print(f"Keine Frequenzen im STFT-Bereich bis {actual_max_freq_display} Hz für {title_info} gefunden.")


def plot_tacho_analysis(
    tacho_data: np.ndarray,
    time_vector: np.ndarray,
    sampling_rate: float,
    expected_rpm_output: float,
    title_info: str,
    n_plot_samples_tacho: int,
    num_pulses_per_rev_output: int = 1 # Annahme, die wir untersuchen wollen
):
    """Analysiert und plottet das Tachosignal."""
    if tacho_data.size == 0 or time_vector.size == 0:
        print(f"Keine Tacho- oder Zeitdaten für {title_info}.")
        return

    # Dynamischer Schwellwert oder robustere Methode könnte hier besser sein
    tacho_min, tacho_max = np.min(tacho_data), np.max(tacho_data)
    if tacho_max == tacho_min: # Konstantes Signal
        tacho_threshold = tacho_max 
        print(f"WARNUNG: Tachosignal für {title_info} ist konstant. Schwellwert bei {tacho_threshold:.2f}. Keine Flanken erwartet.")
    else:
        tacho_threshold = (tacho_max + tacho_min) / 2.0 + (tacho_max - tacho_min) * 0.1 # Etwas über der Mitte

    # Detektiere steigende Flanken
    edges = np.where((tacho_data[:-1] < tacho_threshold) & (tacho_data[1:] >= tacho_threshold))[0] + 1 # +1 für korrekten Index

    print(f"\nAnalyse Tachosignal für {title_info}:")
    print(f"  Anzahl detektierter Tachoimpulse: {len(edges)}")
    
    calculated_output_rpm_from_tacho_corrected = np.nan
    avg_pulse_duration_corrected = np.nan

    if len(edges) > num_pulses_per_rev_output : # Benötigt mindestens N+1 Impulse für N Perioden
        # Zeitpunkte der detektierten Flanken
        edge_timestamps = time_vector[edges]
        
        # Berechne die Dauer zwischen jedem N-ten Impuls (entspricht einer Umdrehung)
        if num_pulses_per_rev_output > 0:
            # Differenzen zwischen Flanken, die num_pulses_per_rev_output auseinander liegen
            revolution_durations = np.diff(edge_timestamps[::num_pulses_per_rev_output])
            if revolution_durations.size > 0:
                avg_pulse_duration_corrected = np.mean(revolution_durations)
                # RPM ist (1 / Dauer einer Umdrehung in s) * 60
                calculated_output_rpm_from_tacho_corrected = (1 / avg_pulse_duration_corrected) * 60
                print(f"  Durchschnittliche Dauer pro {num_pulses_per_rev_output} Impuls(e) (angenommene Umdrehung): {avg_pulse_duration_corrected:.4f} s")
                print(f"  Daraus berechnete Ausgangswellendrehzahl (korrigiert für {num_pulses_per_rev_output}PPR): {calculated_output_rpm_from_tacho_corrected:.2f} RPM")
            else:
                 print(f"  Nicht genügend Impulse für die Berechnung mit {num_pulses_per_rev_output} PPR gefunden.")
        else:
            print("  Ungültige Anzahl von Impulsen pro Umdrehung (num_pulses_per_rev_output muss > 0 sein).")

    print(f"  Erwartete Ausgangswellendrehzahl (aus Eingangs-RPM): {expected_rpm_output:.2f} RPM")

    plt.figure(figsize=(16, 4))
    plt.plot(time_vector[:n_plot_samples_tacho], tacho_data[:n_plot_samples_tacho], label='Tachosignal', linewidth=0.8)
    
    # Detektierte Impulse einzeichnen (nur die im Plot-Bereich)
    plotted_edges = edges[time_vector[edges] <= time_vector[n_plot_samples_tacho-1]] if n_plot_samples_tacho > 0 and time_vector.size >= n_plot_samples_tacho else edges
    plt.plot(time_vector[plotted_edges], tacho_data[plotted_edges], 'rx', markersize=5, label='Detektierte Impulse')
    plt.axhline(tacho_threshold, color='g', linestyle='--', linewidth=1, label=f'Schwelle: {tacho_threshold:.2f}')
    
    plt.title(f"Tachosignal & Detektion - {title_info}", fontsize=10)
    plt.xlabel("Zeit (s)", fontsize=9)
    plt.ylabel("Tacho Amplitude", fontsize=9)
    plt.xticks(fontsize=8)
    plt.yticks(fontsize=8)
    plt.legend(fontsize='x-small')
    plt.grid(True, linestyle=':', alpha=0.7)
    plt.show()
    
    return calculated_output_rpm_from_tacho_corrected, avg_pulse_duration_corrected


# --- Anwendung der Plot-Funktionen auf ausgewählte Beispiele ---

if 'raw_data_df' in locals() and not raw_data_df.empty:
    print("\n--- Visualisierungsbeispiele ---")
    column_names = raw_data_df.columns.tolist()
    display('All of the available columns: ', column_names)
    # Wähle RPM und Torque für konsistente Vergleiche (mittlere Werte als Beispiel)
    unique_rpms = sorted(raw_data_df['rpm_input'].unique())
    target_rpm_example = unique_rpms[len(unique_rpms) // 2] if unique_rpms else 800 # Fallback
    
    unique_torques = sorted(raw_data_df['torque_output_nm'].unique())
    target_torque_example = unique_torques[len(unique_torques) // 2] if unique_torques else 200 # Fallback
    
    print(f"Verwende RPM: {target_rpm_example}, Torque: {target_torque_example} für Vergleichsplots (falls verfügbar).")

    # 1. Gesunder Zustand
    healthy_info = get_representative_file_info(raw_data_df, pitting_level_target=0, rpm_target=target_rpm_example, torque_target=target_torque_example)
    if healthy_info is not None:
        filename_healthy = healthy_info['filename']
        rpm_healthy = healthy_info['rpm_input']
        pitting_lvl_healthy = healthy_info['pitting_level']
        title_healthy = f"Gesund (Lvl {pitting_lvl_healthy}), {filename_healthy}"
        
        data_segment_healthy = raw_data_df[raw_data_df['filename'] == filename_healthy]
        
        VIBRATION_SENSOR_COLUMNS = ['acc_x', 'acc_y', 'acc_z']
        # Rohsignale
        n_plot_samples_time = int(0.1 * SAMPLING_RATE_HZ) # ca. 0.1 Sekunden
        plot_time_signals(data_segment_healthy.head(n_plot_samples_time), VIBRATION_SENSOR_COLUMNS, title_healthy)
        
        # FFT (für acc_z)
        if calculated_frequencies_per_file and filename_healthy in calculated_frequencies_per_file:
            freqs_healthy = calculated_frequencies_per_file[filename_healthy]
            plot_fft_signals(data_segment_healthy, 'acc_z', SAMPLING_RATE_HZ, freqs_healthy, title_healthy, use_log_scale_y=True, max_freq_display=3000)
        
        # STFT (für acc_z)
        plot_stft_signal(data_segment_healthy, 'acc_z', SAMPLING_RATE_HZ, title_healthy, max_freq_display=3000)
        
        # Tachoanalyse
        expected_output_rpm_healthy = rpm_healthy / GEAR_RATIO_INPUT_TO_OUTPUT
        tacho_data_healthy = data_segment_healthy['tacho'].values
        time_vector_healthy = data_segment_healthy['timestamp'].values
        n_plot_samples_tacho = min(len(tacho_data_healthy), int(0.5 * SAMPLING_RATE_HZ)) # 0.5 Sekunden Tacho
        
        # Teste die Hypothese von 6 Impulsen pro Umdrehung
        plot_tacho_analysis(tacho_data_healthy, time_vector_healthy, SAMPLING_RATE_HZ, expected_output_rpm_healthy, title_healthy, n_plot_samples_tacho, num_pulses_per_rev_output=6)

    else:
        print(f"Kein repräsentativer gesunder Datensatz für RPM {target_rpm_example}, Torque {target_torque_example} gefunden.")

    # 2. Geschädigter Zustand (höchster verfügbarer Pitting Level)
    max_pitting_lvl = raw_data_df['pitting_level'].max()
    faulty_info = get_representative_file_info(raw_data_df, pitting_level_target=max_pitting_lvl, rpm_target=target_rpm_example, torque_target=target_torque_example)
    if faulty_info is not None:
        filename_faulty = faulty_info['filename']
        rpm_faulty = faulty_info['rpm_input']
        pitting_lvl_faulty = faulty_info['pitting_level']
        title_faulty = f"Geschädigt (Lvl {pitting_lvl_faulty}), {filename_faulty}"
        
        data_segment_faulty = raw_data_df[raw_data_df['filename'] == filename_faulty]
        
        # Rohsignale
        plot_time_signals(data_segment_faulty.head(n_plot_samples_time), VIBRATION_SENSOR_COLUMNS, title_faulty)
        
        # FFT (für acc_z)
        if calculated_frequencies_per_file and filename_faulty in calculated_frequencies_per_file:
            freqs_faulty = calculated_frequencies_per_file[filename_faulty]
            plot_fft_signals(data_segment_faulty, 'acc_z', SAMPLING_RATE_HZ, freqs_faulty, title_faulty, use_log_scale_y=True, max_freq_display=3000)

        # STFT (für acc_z)
        plot_stft_signal(data_segment_faulty, 'acc_z', SAMPLING_RATE_HZ, title_faulty, max_freq_display=3000)
        
        # Tachoanalyse
        expected_output_rpm_faulty = rpm_faulty / GEAR_RATIO_INPUT_TO_OUTPUT
        tacho_data_faulty = data_segment_faulty['tacho'].values
        time_vector_faulty = data_segment_faulty['timestamp'].values
        n_plot_samples_tacho_faulty = min(len(tacho_data_faulty), int(0.5 * SAMPLING_RATE_HZ))
        plot_tacho_analysis(tacho_data_faulty, time_vector_faulty, SAMPLING_RATE_HZ, expected_output_rpm_faulty, title_faulty, n_plot_samples_tacho_faulty, num_pulses_per_rev_output=6)

    else:
        print(f"Kein repräsentativer geschädigter Datensatz (Lvl {max_pitting_lvl}) für RPM {target_rpm_example}, Torque {target_torque_example} gefunden.")
        
else:
    print("WARNUNG: DataFrame `raw_data_df` ist leer. Visualisierungen können nicht erstellt werden.")
```

```python
# Weiter in Sektion II: Datenexploration und -vorverarbeitung

# Markdown Zelle: Einleitung Merkmalsextraktion
"""
### 2.4 Merkmalsextraktion (Feature Engineering) 🛠️

Um die Rohdaten für Machine-Learning-Modelle aufzubereiten, extrahieren wir aussagekräftige Merkmale.
Wir werden für jede einzelne Messdatei (d.h. jede .txt-Datei) einen Satz von statistischen Merkmalen
im Zeit- und Frequenzbereich berechnen. Dies transformiert die langen Zeitreihen in kompakte
Merkmalsvektoren, die den Zustand des Getriebes charakterisieren.
"""

# Code Zelle: Merkmalsextraktionsfunktionen

from scipy.stats import skew, kurtosis
# `librosa` könnte für fortgeschrittenere Audio-/Vibrationsmerkmale nützlich sein,
# aber für den Anfang reichen scipy/numpy.
# import librosa 

def get_time_domain_features(segment: pd.Series) -> dict:
    """
    Berechnet statistische Merkmale im Zeitbereich für ein gegebenes Signal-Segment.
    """
    features = {}
    # Sicherstellen, dass das Segment nicht leer ist und numerische Daten enthält
    if segment.empty or not pd.api.types.is_numeric_dtype(segment):
        # Standardwerte für leere oder nicht-numerische Segmente zurückgeben
        keys = ['mean', 'std', 'rms', 'skewness', 'kurt', 'peak', 'peak_to_peak', 'crest_factor', 'shape_factor', 'clearance_factor']
        for k in keys:
            features[k] = np.nan
        return features

    data = segment.values # Konvertiere zu NumPy Array für Performance

    features['mean'] = np.mean(data)
    features['std'] = np.std(data)
    features['rms'] = np.sqrt(np.mean(data**2))
    features['skewness'] = skew(data)
    features['kurt'] = kurtosis(data) # Exzess-Kurtosis (Normalverteilung hat Kurtosis 0)
    
    peak_val = np.max(np.abs(data))
    features['peak'] = peak_val
    features['peak_to_peak'] = np.max(data) - np.min(data)
    
    if features['rms'] != 0:
        features['crest_factor'] = peak_val / features['rms']
    else:
        features['crest_factor'] = np.nan # Vermeide Division durch Null

    mean_abs = np.mean(np.abs(data))
    if mean_abs != 0:
        features['shape_factor'] = features['rms'] / mean_abs
        features['clearance_factor'] = peak_val / (np.mean(np.sqrt(np.abs(data)))**2) if np.mean(np.sqrt(np.abs(data))) != 0 else np.nan
    else:
        features['shape_factor'] = np.nan
        features['clearance_factor'] = np.nan
        
    return features


def get_frequency_domain_features(
    segment: pd.Series,
    sampling_rate: float,
    char_frequencies_for_segment: dict = None # Optional: charakteristische Frequenzen für Bandenergie
) -> dict:
    """
    Berechnet Merkmale im Frequenzbereich für ein gegebenes Signal-Segment.
    """
    features = {}
    N = len(segment)
    if N == 0 or not pd.api.types.is_numeric_dtype(segment):
        keys = ['spectral_centroid', 'spectral_std', 'spectral_skewness', 'spectral_kurtosis', 'gmf_band_energy_rel']
        for k in keys:
            features[k] = np.nan
        return features

    data = segment.values
    
    yf = fft(data)
    # Einseitiges Spektrum, nur positive Frequenzen
    magnitude_spectrum = 2.0/N * np.abs(yf[0:N//2])
    frequencies = fftfreq(N, 1/sampling_rate)[0:N//2]

    if frequencies.size == 0: # Falls N < 2
        # Standardwerte zurückgeben, falls keine Frequenzen berechnet werden können
        keys = ['spectral_centroid', 'spectral_std', 'spectral_skewness', 'spectral_kurtosis', 'gmf_band_energy_rel']
        for k in keys:
            features[k] = np.nan
        return features

    # Vermeide Division durch Null, wenn das gesamte Spektrum Null ist
    if np.sum(magnitude_spectrum) == 0:
        features['spectral_centroid'] = 0.0
        features['spectral_std'] = 0.0
        features['spectral_skewness'] = 0.0
        features['spectral_kurtosis'] = 0.0
    else:
        # Spektraler Mittelwert (Centroid)
        features['spectral_centroid'] = np.sum(frequencies * magnitude_spectrum) / np.sum(magnitude_spectrum)
        # Spektrale Standardabweichung, Schiefe, Kurtosis
        # Diese sind komplexer und erfordern eine sorgfältige Definition als gewichtete Momente
        # Hier eine vereinfachte Form für std; skew/kurtosis werden oft über spezialisierte Bibliotheken gemacht
        features['spectral_std'] = np.sqrt(np.sum(((frequencies - features['spectral_centroid'])**2) * magnitude_spectrum) / np.sum(magnitude_spectrum))
        
        # Für spektrale Schiefe und Kurtosis könnten wir die Definitionen implementieren,
        # oder sie vorerst weglassen, da ihre Implementierung fehleranfällig sein kann.
        # Für eine Bewerbung ist es okay, sich auf die wichtigsten und robust implementierbaren zu konzentrieren.
        features['spectral_skewness'] = np.nan # TODO: Implement or use library
        features['spectral_kurtosis'] = np.nan # TODO: Implement or use library

    # Energie in bestimmten Bändern (Beispiel: um GMF)
    total_energy = np.sum(magnitude_spectrum**2)
    features['gmf_band_energy_rel'] = np.nan # Standardwert

    if char_frequencies_for_segment and 'gmf_hz' in char_frequencies_for_segment and total_energy > 0:
        gmf = char_frequencies_for_segment['gmf_hz']
        # Definiere ein Band um die GMF, z.B. GMF +/- 5% oder eine feste Bandbreite
        band_width_hz = max(10, 0.05 * gmf) # Mindestens 10 Hz oder 5% der GMF
        lower_bound = gmf - band_width_hz / 2
        upper_bound = gmf + band_width_hz / 2
        
        idx_in_band = np.where((frequencies >= lower_bound) & (frequencies <= upper_bound))[0]
        if idx_in_band.size > 0:
            gmf_band_energy = np.sum(magnitude_spectrum[idx_in_band]**2)
            features['gmf_band_energy_rel'] = gmf_band_energy / total_energy
            
    return features

# --- Anwendung der Merkmalsextraktion auf den DataFrame ---
# Wir gruppieren nach 'filename', um für jede Datei Merkmale zu extrahieren.

# Sicherstellen, dass die benötigten globalen Variablen existieren
if 'raw_data_df' not in locals() or raw_data_df.empty:
    print("FEHLER: `raw_data_df` ist nicht verfügbar. Merkmalsextraktion kann nicht durchgeführt werden.")
    # Hier könnte man das Notebook stoppen oder eine Exception werfen
    # raise ValueError("`raw_data_df` nicht gefunden oder leer.")
    # Für jetzt wird ein leerer features_df erstellt, wenn raw_data_df nicht existiert
    features_list = []
    feature_columns_base = ['filename', 'pitting_level', 'rpm_input', 'torque_output_nm']
else:
    print(f"\nStarte Merkmalsextraktion für {raw_data_df['filename'].nunique()} Dateien...")
    # `calculated_frequencies_per_file` sollte aus dem vorherigen Schritt verfügbar sein
    if 'calculated_frequencies_per_file' not in locals():
        print("WARNUNG: `calculated_frequencies_per_file` nicht gefunden. Frequenzbasierte Merkmale sind möglicherweise unvollständig.")
        calculated_frequencies_per_file = {} # Leeres Dict, um Fehler zu vermeiden

    features_list = []
    # Gruppiere nach Dateinamen, um jede Datei als ein Sample zu behandeln
    # `grouped = raw_data_df.groupby('filename')` ist sehr speicherintensiv bei 260Mio Zeilen.
    # Besser: Iteriere über einzigartige Dateinamen und filtere den DataFrame.

    unique_filenames = raw_data_df['filename'].unique()
    
    # Fortschrittsanzeige (einfach)
    total_files = len(unique_filenames)
    processed_count = 0

    for filename in unique_filenames:
        # Hole das Segment (alle Zeilen) für die aktuelle Datei
        file_segment_df = raw_data_df[raw_data_df['filename'] == filename]
        
        if file_segment_df.empty:
            continue

        # Metadaten für diese Datei (bleiben für alle Sensoren gleich)
        # Nehme Metadaten von der ersten Zeile des Segments
        meta_row = file_segment_df.iloc[0]
        current_features = {
            'filename': filename,
            'pitting_level': meta_row['pitting_level'],
            'rpm_input': meta_row['rpm_input'],
            'torque_output_nm': meta_row['torque_output_nm']
            # Füge hier ggf. 'repetition' hinzu, falls es als Merkmal dienen soll
        }
        
        char_freqs_for_file = calculated_frequencies_per_file.get(filename, {})

        # Merkmale für jeden Vibrationssensor extrahieren
        for sensor_col in VIBRATION_SENSOR_COLUMNS: # ['acc_x', 'acc_y', 'acc_z']
            sensor_segment = file_segment_df[sensor_col]
            
            time_feats = get_time_domain_features(sensor_segment)
            for key, value in time_feats.items():
                current_features[f'{sensor_col}_time_{key}'] = value
            
            freq_feats = get_frequency_domain_features(sensor_segment, SAMPLING_RATE_HZ, char_freqs_for_file)
            for key, value in freq_feats.items():
                current_features[f'{sensor_col}_freq_{key}'] = value
        
        features_list.append(current_features)
        
        processed_count += 1
        if processed_count % 50 == 0 or processed_count == total_files: # Alle 50 Dateien oder am Ende eine Meldung
            print(f"  Verarbeitet: {processed_count}/{total_files} Dateien ({processed_count/total_files*100:.1f}%)")


# Erstelle den finalen DataFrame mit den extrahierten Merkmalen
features_df = pd.DataFrame(features_list)

if not features_df.empty:
    print("\nMerkmalsextraktion abgeschlossen.")
    print(f"Shape des Merkmals-DataFrames: {features_df.shape}")
    print("Beispiel-Merkmale (erste 5 Zeilen):")
    print(features_df.head())

    # Überprüfe auf NaN-Werte im Merkmals-DataFrame
    nan_counts = features_df.isnull().sum()
    nan_cols = nan_counts[nan_counts > 0]
    if not nan_cols.empty:
        print("\nSpalten mit NaN-Werten im Merkmals-DataFrame:")
        print(nan_cols)
        print("Diese könnten durch leere Segmente, Division durch Null oder fehlende Frequenzen entstanden sein.")
        # TODO: Überlege dir eine Strategie zur Behandlung von NaNs (z.B. Imputation mit Mittelwert/Median, oder Entfernen der Zeilen/Spalten)
    else:
        print("\nKeine NaN-Werte im Merkmals-DataFrame gefunden.")
else:
    print("WARNUNG: Der Merkmals-DataFrame `features_df` ist leer. Überprüfe die Extraktionslogik und die Eingabedaten.")


# Speichere den Merkmals-DataFrame für spätere Verwendung (optional, aber empfohlen)
if not features_df.empty:
    FEATURES_PARQUET_FILE_PATH = os.path.join(PROCESSED_DATA_DIR, "phm2023_na_features.parquet")
    try:
        features_df.to_parquet(FEATURES_PARQUET_FILE_PATH, index=False)
        print(f"\nMerkmals-DataFrame erfolgreich als Parquet gespeichert unter: {FEATURES_PARQUET_FILE_PATH}")
    except Exception as e:
        print(f"FEHLER beim Speichern des Merkmals-DataFrames: {e}")
```

```python
# Code Zelle: Feature-Extraktion mit Caching-Logik

import os
import time
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
from scipy.fft import fft, fftfreq

# Annahmen:
# - raw_data_df: DataFrame mit Rohdaten ist verfügbar
# - calculated_frequencies_per_file: Dictionary mit Frequenzen ist verfügbar
# - VIBRATION_SENSOR_COLUMNS = ['acc_x', 'acc_y', 'acc_z']
# - SAMPLING_RATE_HZ = 20480
# - PROCESSED_DATA_DIR = "./processed_data"

# --- Pfad zur zwischengespeicherten Feature-Datei ---
# Definiere den Dateinamen für die endgültigen Features.
FEATURES_PARQUET_PATH = os.path.join(PROCESSED_DATA_DIR, "phm2023_na_features_final.parquet")

# --- Überprüfe, ob die Feature-Datei bereits existiert ---
if os.path.exists(FEATURES_PARQUET_PATH):
    print(f"Zwischengespeicherte Feature-Datei gefunden. Lade direkt aus: {FEATURES_PARQUET_PATH}")
    try:
        # Lade die bereits berechneten Features. Dies ist der schnelle Pfad.
        features_df = pd.read_parquet(FEATURES_PARQUET_PATH)
        print("Merkmals-DataFrame erfolgreich geladen.")
        print(f"Shape: {features_df.shape}")
        display(features_df.head())
    except Exception as e:
        print(f"FEHLER beim Laden der zwischengespeicherten Datei: {e}")
        print("Setze fort mit der Neuberechnung der Features.")
        features_df = pd.DataFrame() # Setze zurück, um Neuberechnung zu erzwingen
else:
    print(f"Keine zwischengespeicherte Feature-Datei unter {FEATURES_PARQUET_PATH} gefunden.")
    print("Starte die Merkmalsextraktion (dies kann einige Zeit dauern)...")
    features_df = pd.DataFrame() # Initialisiere, falls die Neuberechnung fehlschlägt

# --- Führe die Merkmalsextraktion nur aus, wenn der DataFrame nicht geladen wurde ---
if 'features_df' not in locals() or features_df.empty:
    
    # Stelle sicher, dass die Basisdaten vorhanden sind
    if 'raw_data_df' in locals() and not raw_data_df.empty:

        # Deine Feature-Extraktionsfunktionen (hier für Vollständigkeit enthalten)
        def get_time_domain_features(segment: pd.Series) -> dict:
            features = {}
            keys = ['mean', 'std', 'rms', 'skewness', 'kurt', 'peak', 'peak_to_peak', 'crest_factor', 'shape_factor', 'clearance_factor']
            for k in keys: features[k] = np.nan
            if segment.empty: return features
            data = segment.values
            features['mean'] = np.mean(data)
            features['std'] = np.std(data)
            features['rms'] = np.sqrt(np.mean(data**2))
            features['skewness'] = skew(data)
            features['kurt'] = kurtosis(data)
            peak_val = np.max(np.abs(data))
            features['peak'] = peak_val
            features['peak_to_peak'] = np.max(data) - np.min(data)
            if features['rms'] != 0: features['crest_factor'] = peak_val / features['rms']
            mean_abs = np.mean(np.abs(data))
            if mean_abs != 0:
                features['shape_factor'] = features['rms'] / mean_abs
                features['clearance_factor'] = peak_val / (np.mean(np.sqrt(np.abs(data)))**2) if np.mean(np.sqrt(np.abs(data))) != 0 else np.nan
            return features

        def get_frequency_domain_features(segment: pd.Series, sampling_rate: float, char_frequencies_for_segment: dict = None) -> dict:
            features = {}
            N = len(segment)
            keys = ['spectral_centroid', 'spectral_std', 'gmf_band_energy_rel']
            for k in keys: features[k] = np.nan
            if N < 2: return features
            data = segment.values
            yf = fft(data)
            magnitude_spectrum = 2.0/N * np.abs(yf[0:N//2])
            frequencies = fftfreq(N, 1/sampling_rate)[0:N//2]
            if frequencies.size == 0: return features
            if np.sum(magnitude_spectrum) == 0:
                features['spectral_centroid'], features['spectral_std'] = 0.0, 0.0
                return features
            features['spectral_centroid'] = np.sum(frequencies * magnitude_spectrum) / np.sum(magnitude_spectrum)
            features['spectral_std'] = np.sqrt(np.sum(((frequencies - features['spectral_centroid'])**2) * magnitude_spectrum) / np.sum(magnitude_spectrum))
            total_energy = np.sum(magnitude_spectrum**2)
            if char_frequencies_for_segment and 'gmf_hz' in char_frequencies_for_segment and total_energy > 0:
                gmf = char_frequencies_for_segment['gmf_hz']
                band_width_hz = max(10, 0.05 * gmf)
                idx_in_band = np.where((frequencies >= gmf - band_width_hz / 2) & (frequencies <= gmf + band_width_hz / 2))[0]
                if idx_in_band.size > 0:
                    gmf_band_energy = np.sum(magnitude_spectrum[idx_in_band]**2)
                    features['gmf_band_energy_rel'] = gmf_band_energy / total_energy
            return features

        # Sequentielle Extraktionsschleife (dein Originalcode)
        features_list = []
        unique_filenames = raw_data_df['filename'].unique()
        total_files = len(unique_filenames)
        start_time = time.time()

        for i, filename in enumerate(unique_filenames):
            file_segment_df = raw_data_df[raw_data_df['filename'] == filename]
            if file_segment_df.empty: continue
            
            meta_row = file_segment_df.iloc[0]
            current_features = {
                'filename': filename,
                'pitting_level': meta_row['pitting_level'],
                'rpm_input': meta_row['rpm_input'],
                'torque_output_nm': meta_row['torque_output_nm']
            }
            char_freqs_for_file = calculated_frequencies_per_file.get(filename, {})
            
            for sensor_col in VIBRATION_SENSOR_COLUMNS:
                sensor_segment = file_segment_df[sensor_col]
                time_feats = get_time_domain_features(sensor_segment)
                for key, value in time_feats.items():
                    current_features[f'{sensor_col}_time_{key}'] = value
                freq_feats = get_frequency_domain_features(sensor_segment, SAMPLING_RATE_HZ, char_freqs_for_file)
                for key, value in freq_feats.items():
                    current_features[f'{sensor_col}_freq_{key}'] = value
            features_list.append(current_features)
            
            # Fortschrittsanzeige
            if (i + 1) % 50 == 0 or (i + 1) == total_files:
                print(f"  Verarbeitet: {i + 1}/{total_files} Dateien ({(i + 1)/total_files*100:.1f}%)")

        features_df = pd.DataFrame(features_list)
        end_time = time.time()
        print(f"\nMerkmalsextraktion abgeschlossen in {end_time - start_time:.2f} Sekunden.")

        # --- Speichere die neu berechneten Features für das nächste Mal ---
        if not features_df.empty:
            print(f"Shape des neuen Merkmals-DataFrames: {features_df.shape}")
            try:
                # Stelle sicher, dass das Verzeichnis existiert
                os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)
                features_df.to_parquet(FEATURES_PARQUET_PATH, index=False)
                print(f"Merkmals-DataFrame erfolgreich zwischengespeichert unter: {FEATURES_PARQUET_PATH}")
            except Exception as e:
                print(f"FEHLER beim Speichern des neuen Merkmals-DataFrames: {e}")
        else:
            print("WARNUNG: Der neu erstellte Merkmals-DataFrame ist leer.")

    else:
        print("FEHLER: Rohdaten-DataFrame (`raw_data_df`) ist nicht verfügbar, um Features neu zu berechnen.")

# Finale Überprüfung, ob 'features_df' jetzt existiert und nicht leer ist.
if 'features_df' in locals() and not features_df.empty:
    print("\n`features_df` ist nun für die weiteren Schritte bereit.")
else:
    print("\nFEHLER: Konnte `features_df` weder laden noch neu erstellen.")
```

```python
# Code Zelle: Automatisierte und Parallelisierte Merkmalsextraktion

import os
import time
import pandas as pd
import numpy as np
import concurrent.futures
import multiprocessing

# 1. Importiere die Worker-Funktion aus der externen .py Datei
# Dies ist der entscheidende Schritt, um den `AttributeError` zu beheben.
try:
    from feature_extractor_utils import process_single_file_features_external
    print("Worker-Funktion 'process_single_file_features_external' erfolgreich importiert.")
except ImportError:
    print("FEHLER: 'feature_extractor_utils.py' konnte nicht gefunden werden. Stelle sicher, dass die Datei im selben Verzeichnis wie dieses Notebook liegt.")
    # Stoppe die Ausführung dieser Zelle, wenn der Import fehlschlägt.
    # raise


# 2. Hauptfunktion zur Orchestrierung der Parallelisierung
def automated_parallel_feature_extraction(
    raw_df: pd.DataFrame,
    vibration_sensor_cols: list,
    sampling_rate_hz: float,
    all_char_freqs_per_file: dict,
    num_workers: int = None
) -> pd.DataFrame:
    """
    Führt die Merkmalsextraktion für alle Dateien im raw_df parallel durch.
    """
    if raw_df.empty:
        print("Eingabe-DataFrame ist leer. Keine Merkmale zu extrahieren.")
        return pd.DataFrame()

    unique_filenames_list = raw_df['filename'].unique()
    total_files = len(unique_filenames_list)
    print(f"\nStarte parallelisierte Merkmalsextraktion für {total_files} einzigartige Dateien...")

    if num_workers is None:
        # Heuristik: Nutze die meisten CPUs, aber begrenze auf z.B. 8, um das System nicht zu überlasten.
        # Passe dies bei Bedarf an deine Hardware an.
        num_workers = min(multiprocessing.cpu_count(), 8)
        if total_files < num_workers: # Nicht mehr Worker als Aufgaben
            num_workers = total_files
    print(f"Verwende {num_workers} Worker-Prozesse.")

    features_list_parallel = []
    start_time = time.time()

    # Die 'with'-Anweisung stellt sicher, dass der Pool korrekt verwaltet wird.
    with concurrent.futures.ProcessPoolExecutor(max_workers=num_workers) as executor:
        # Erstelle eine Map von Future-Objekten zu Dateinamen für eine einfache Nachverfolgung
        futures = {
            executor.submit(
                process_single_file_features_external, # Die importierte Funktion
                fname,
                raw_df, 
                vibration_sensor_cols,
                sampling_rate_hz,
                all_char_freqs_per_file.get(fname, {}) # Frequenzen für diese Datei
            ): fname for fname in unique_filenames_list
        }

        processed_count = 0
        for future in concurrent.futures.as_completed(futures):
            filename_completed = futures[future]
            try:
                result = future.result() # Holt das Ergebnis (oder die Ausnahme) der Worker-Funktion
                if result:
                    if 'error' in result:
                        print(f"Datei {filename_completed} hatte einen Fehler: {result['error']}")
                    else:
                        features_list_parallel.append(result)
                else: # Falls der Worker None zurückgibt
                    print(f"Kein Ergebnis für Datei {filename_completed} (möglicherweise leer oder übersprungen).")
            except Exception as exc:
                print(f"Datei {filename_completed} hat eine unerwartete Ausnahme im Hauptprozess erzeugt: {exc}")
            
            processed_count += 1
            # Fortschrittsanzeige
            if processed_count % (total_files // 20 if total_files > 40 else 2) == 0 or processed_count == total_files:
                elapsed_time_so_far = time.time() - start_time
                print(f"  Verarbeitet: {processed_count}/{total_files} Dateien ({processed_count/total_files*100:.1f}%) - Zeit: {elapsed_time_so_far:.2f}s")

    end_time = time.time()
    print(f"\nParallelisierte Merkmalsextraktion abgeschlossen in {end_time - start_time:.2f} Sekunden.")

    final_features_df = pd.DataFrame(features_list_parallel)
    return final_features_df


# 3. Anwendung der automatisierten und parallelisierten Funktion
# Annahme: 'raw_data_df' und 'calculated_frequencies_per_file' sind aus vorherigen Zellen verfügbar.

if 'raw_data_df' in locals() and not raw_data_df.empty:
    # Fallback, falls das Frequenz-Dictionary aus irgendeinem Grund nicht existiert
    if 'calculated_frequencies_per_file' not in locals() or not calculated_frequencies_per_file:
        print("WARNUNG: `calculated_frequencies_per_file` nicht verfügbar. Erstelle es neu...")
        calculated_frequencies_per_file = {}
        if 'filename' in raw_data_df.columns and 'rpm_input' in raw_data_df.columns:
            # Effizienter Weg, um Frequenzen für jede einzigartige Datei zu berechnen
            unique_files_info = raw_data_df[['filename', 'rpm_input']].drop_duplicates()
            for _, row in unique_files_info.iterrows():
                calculated_frequencies_per_file[row['filename']] = calculate_gear_characteristic_frequencies(row['rpm_input'])
            print(f"{len(calculated_frequencies_per_file)} Frequenz-Sets erstellt.")

    # Führe die parallelisierte Extraktion durch
    features_df_parallel = automated_parallel_feature_extraction(
        raw_df=raw_data_df,
        vibration_sensor_cols=VIBRATION_SENSOR_COLUMNS,
        sampling_rate_hz=SAMPLING_RATE_HZ,
        all_char_freqs_per_file=calculated_frequencies_per_file,
        num_workers=None # Lässt die Funktion die Anzahl der Worker bestimmen
    )

    if not features_df_parallel.empty:
        # Ordne die Spalten in einer logischen Reihenfolge (Metadaten zuerst, dann Features)
        meta_cols = ['filename', 'pitting_level', 'rpm_input', 'torque_output_nm', 'repetition']
        feature_cols_sorted = sorted([col for col in features_df_parallel.columns if col not in meta_cols])
        features_df_parallel = features_df_parallel[meta_cols + feature_cols_sorted]
        
        print(f"\nShape des parallel extrahierten Merkmals-DataFrames: {features_df_parallel.shape}")
        print("Beispiel-Merkmale (erste 5 Zeilen):")
        display(features_df_parallel.head()) # display() ist oft schöner im Notebook

        # Speichern des parallel extrahierten Feature-DataFrames
        FEATURES_PARQUET_FILE_PATH = os.path.join(PROCESSED_DATA_DIR, "phm2023_na_features_parallel.parquet")
        try:
            features_df_parallel.to_parquet(FEATURES_PARQUET_FILE_PATH, index=False)
            print(f"\nParallel extrahierter Merkmals-DataFrame erfolgreich gespeichert: {FEATURES_PARQUET_FILE_PATH}")
        except Exception as e:
            print(f"FEHLER beim Speichern des parallel extrahierten Merkmals-DataFrames: {e}")
    else:
        print("WARNUNG: Der parallel extrahierte Merkmals-DataFrame ist nach der Verarbeitung leer.")
else:
    print("FEHLER: `raw_data_df` ist nicht verfügbar. Automatisierte Merkmalsextraktion übersprungen.")
```

```python
# Weiter im Notebook, nach der parallelisierten Merkmalsextraktion und dem Laden von features_df_parallel

# Annahme: features_df_parallel ist dein DataFrame mit den extrahierten Merkmalen
# Wenn du mit 'features_df' weiterarbeiten möchtest:
if 'features_df_parallel' in locals() and not features_df_parallel.empty:
    features_df = features_df_parallel.copy()
    print(f"DataFrame `features_df_parallel` ({features_df.shape}) wurde nach `features_df` kopiert.")
elif 'features_df' not in locals() or features_df.empty:
    print("FEHLER: Weder `features_df_parallel` noch `features_df` sind gültige DataFrames für die Bereinigung.")
    # raise ValueError("Kein gültiger Features-DataFrame zum Arbeiten vorhanden.")
    # Für den weiteren Ablauf wird ein leerer DataFrame erstellt, wenn nichts geladen wurde
    features_df = pd.DataFrame() 

# Markdown Zelle: Einleitung Datenbereinigung und -aufteilung
"""
### 2.5 Datenbereinigung und -aufteilung

Nach der Merkmalsextraktion bereinigen wir den resultierenden DataFrame und teilen ihn
in Trainings-, Validierungs- und Testsets auf, um unsere Modelle fair bewerten zu können.
Wir berücksichtigen dabei die Vorgabe, dass bestimmte Schädigungslevel für das Training
ausgelassen werden sollen.
"""

# Code Zelle: NaN-Behandlung
if not features_df.empty:
    print("\n--- 2.5.1 NaN-Behandlung ---")
    print(f"Ursprünglicher Shape des features_df: {features_df.shape}")
    nan_info_before = features_df.isnull().sum()
    nan_cols_before = nan_info_before[nan_info_before > 0].sort_values(ascending=False)
    
    if not nan_cols_before.empty:
        print("\nSpalten mit NaN-Werten (Anzahl NaNs):")
        print(nan_cols_before)
        
        # Strategie:
        # 1. Spalten entfernen, die einen sehr hohen Anteil an NaNs haben (z.B. > 50%)
        threshold_col_removal = 0.5 * len(features_df)
        cols_to_drop = nan_cols_before[nan_cols_before > threshold_col_removal].index.tolist()
        if cols_to_drop:
            features_df.drop(columns=cols_to_drop, inplace=True)
            print(f"\nSpalten entfernt (mehr als {threshold_col_removal*100/len(features_df):.0f}% NaNs): {cols_to_drop}")
            print(f"Shape nach Spaltenentfernung: {features_df.shape}")
        
        # 2. Zeilen entfernen, wenn eine Zeile (Messung/Datei) zu viele NaNs hat (nach Spaltenentfernung)
        # Dies ist hier weniger wahrscheinlich, da Features pro Datei berechnet wurden.
        # features_df.dropna(axis=0, thresh=features_df.shape[1] * 0.8, inplace=True) # Behalte Zeilen mit mind. 80% Nicht-NaNs
        # print(f"Shape nach Zeilenentfernung (mit vielen NaNs): {features_df.shape}")

        # 3. Imputation für verbleibende NaNs (z.B. mit Median)
        # Wir trennen numerische und nicht-numerische Spalten für die Imputation
        numeric_cols = features_df.select_dtypes(include=np.number).columns.tolist()
        # Identifiziere Spalten, die tatsächlich imputiert werden müssen (numerisch und mit NaNs)
        numeric_cols_with_nan = features_df[numeric_cols].isnull().sum()
        numeric_cols_to_impute = numeric_cols_with_nan[numeric_cols_with_nan > 0].index.tolist()

        if numeric_cols_to_impute:
            from sklearn.impute import SimpleImputer
            imputer = SimpleImputer(strategy='median') # Median ist robuster gegenüber Ausreißern
            
            # Wichtig: Imputer nur auf numerische Spalten anwenden, die NaNs haben
            features_df[numeric_cols_to_impute] = imputer.fit_transform(features_df[numeric_cols_to_impute])
            print(f"\nNaN-Werte in {len(numeric_cols_to_impute)} numerischen Spalten mit Median imputiert.")
            # Überprüfe, ob es immer noch NaNs gibt
            if features_df[numeric_cols_to_impute].isnull().sum().sum() > 0:
                 print("WARNUNG: Immer noch NaNs nach der Imputation vorhanden in numerischen Spalten!")
        else:
            print("\nKeine numerischen Spalten mit NaNs zur Imputation gefunden (nach evtl. Spaltenentfernung).")
            
        nan_info_after = features_df.isnull().sum().sum()
        if nan_info_after == 0:
            print("\nAlle NaN-Werte erfolgreich behandelt.")
        else:
            print(f"\nWARNUNG: Es sind noch {nan_info_after} NaN-Werte im DataFrame vorhanden. Erneute Prüfung notwendig.")
            print(features_df.isnull().sum()[features_df.isnull().sum() > 0])

    else:
        print("Keine NaN-Werte im ursprünglichen features_df gefunden.")
else:
    print("features_df ist leer. NaN-Behandlung übersprungen.")


# Code Zelle: Zielvariable und Aufteilungsvorbereitung
if not features_df.empty:
    print("\n--- 2.5.2 Zielvariable und Aufteilungsvorbereitung ---")
    TARGET_COLUMN = 'pitting_level'
    
    if TARGET_COLUMN not in features_df.columns:
        print(f"FEHLER: Zielspalte '{TARGET_COLUMN}' nicht im features_df gefunden!")
        # Hier abbrechen oder Dummy-Zielvariable für Testzwecke erstellen
    else:
        print(f"Zielvariable: '{TARGET_COLUMN}'")
        print(f"Verteilung der Zielvariablen:\n{features_df[TARGET_COLUMN].value_counts().sort_index()}")

        # Definition der Level, die für das Training ausgelassen werden sollen (gemäß Recherche)
        LEVELS_TO_EXCLUDE_FROM_TRAIN = [5, 7, 9, 10] # Und ggf. weitere, wenn sie als reines Testset dienen
        print(f"Level, die aus Trainingsdaten ausgeschlossen werden (als Testset dienen): {LEVELS_TO_EXCLUDE_FROM_TRAIN}")

        # Trenne die expliziten Testdaten (ausgelassene Level) vom Rest
        test_set_explicit_df = features_df[features_df[TARGET_COLUMN].isin(LEVELS_TO_EXCLUDE_FROM_TRAIN)]
        remaining_data_df = features_df[~features_df[TARGET_COLUMN].isin(LEVELS_TO_EXCLUDE_FROM_TRAIN)]

        print(f"Shape des expliziten Testsets (Level {LEVELS_TO_EXCLUDE_FROM_TRAIN}): {test_set_explicit_df.shape}")
        print(f"Shape der verbleibenden Daten (für Training/Validierung): {remaining_data_df.shape}")

        if remaining_data_df.empty:
            print("FEHLER: Nachdem die expliziten Testlevel entfernt wurden, bleiben keine Daten für Training/Validierung übrig!")
        else:
            # Merkmale (X) und Ziel (y) für die verbleibenden Daten
            X = remaining_data_df.drop(columns=[TARGET_COLUMN, 'filename']) # 'filename' ist keine Merkmalsspalte
            y = remaining_data_df[TARGET_COLUMN]
            
            # Merkmale (X_test_expl) und Ziel (y_test_expl) für das explizite Testset
            X_test_explicit = pd.DataFrame()
            y_test_explicit = pd.Series(dtype=y.dtype) 
            if not test_set_explicit_df.empty:
                 X_test_explicit = test_set_explicit_df.drop(columns=[TARGET_COLUMN, 'filename'])
                 y_test_explicit = test_set_explicit_df[TARGET_COLUMN]
            
            print(f"X (Train/Valid) Shape: {X.shape}, y (Train/Valid) Shape: {y.shape}")
            if not X_test_explicit.empty:
                print(f"X_test_explicit Shape: {X_test_explicit.shape}, y_test_explicit Shape: {y_test_explicit.shape}")
            else:
                print("Das explizite Testset ist leer (keine der Ausschluss-Level waren in den Daten).")


            # Aufteilung der verbleibenden Daten in Trainings- und Validierungssets
            # (falls remaining_data_df nicht leer ist)
            from sklearn.model_selection import train_test_split

            # TODO: Entscheide, ob du eine Klassifikations- oder Regressionsaufgabe hast.
            #       Für Stratifizierung bei Klassifikation muss y diskret sein.
            #       Wenn y kontinuierlich (oder viele Level) ist und du stratifizieren willst,
            #       müsstest du y möglicherweise erst binnnen (in Klassen einteilen).
            # Annahme hier: Regression, daher keine Stratifizierung auf y direkt,
            # aber wir versuchen, nach `rpm_input` und `torque_output_nm` zu stratifizieren, wenn möglich,
            # oder zumindest sicherstellen, dass die Verteilung ähnlich ist.
            # Einfacher Ansatz: Zufälliger Split, wenn keine komplexere Gruppenstruktur berücksichtigt wird.
            
            stratify_cols_train_val = None # Standardmäßig keine Stratifizierung
            # Falls 'pitting_level' für die Stratifizierung im remaining_data_df genutzt werden soll
            # und es genügend Samples pro Klasse gibt:
            # if y.nunique() < len(y) * 0.5 and y.nunique() > 1 : # Heuristik für diskrete Klassen
            #     value_counts = y.value_counts()
            #     if (value_counts >= 2).all(): # Mindestens 2 Samples pro Klasse für Stratifizierung
            #        stratify_cols_train_val = y
            #        print(f"Stratifiziere Train/Val-Split nach '{TARGET_COLUMN}'.")
            #     else:
            #        print(f"Nicht genügend Samples in jeder Klasse von '{TARGET_COLUMN}' für Stratifizierung. Führe zufälligen Split durch.")
            # else:
            #    print(f"'{TARGET_COLUMN}' scheint nicht für Stratifizierung geeignet (zu viele unique Werte oder nur ein Wert). Führe zufälligen Split durch.")


            if len(X) > 1: # Nur splitten, wenn es mehr als einen Datensatz gibt
                X_train, X_val, y_train, y_val = train_test_split(
                    X, y,
                    test_size=0.20, # z.B. 20% für Validierung
                    random_state=RANDOM_SEED, # Für Reproduzierbarkeit
                    stratify=stratify_cols_train_val # Kann None sein
                )
                print(f"Trainingsset: X_train Shape {X_train.shape}, y_train Shape {y_train.shape}")
                print(f"Validierungsset: X_val Shape {X_val.shape}, y_val Shape {y_val.shape}")

                # Merkmalskalierung (StandardScaler)
                # WICHTIG: Scaler auf Trainingsdaten anpassen, dann auf Validierungs- und Testdaten anwenden!
                from sklearn.preprocessing import StandardScaler
                
                # Identifiziere numerische Spalten, die skaliert werden sollen
                # (schließe ggf. bereits kodierte kategorische oder binäre Spalten aus, falls vorhanden)
                cols_to_scale = X_train.select_dtypes(include=np.number).columns.tolist()
                # Entferne Spalten, die möglicherweise keine Skalierung benötigen (z.B. 'repetition', wenn es als Merkmal behalten wurde)
                # oder bereits Indikatorspalten sind.
                # Hier nehmen wir an, alle numerischen Spalten in X_train (außer ggf. 'repetition') sollen skaliert werden.
                # if 'repetition' in cols_to_scale:
                #     cols_to_scale.remove('repetition')

                if cols_to_scale:
                    scaler = StandardScaler()
                    X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
                    X_val[cols_to_scale] = scaler.transform(X_val[cols_to_scale])
                    if not X_test_explicit.empty and all(col in X_test_explicit.columns for col in cols_to_scale):
                         X_test_explicit[cols_to_scale] = scaler.transform(X_test_explicit[cols_to_scale])
                    
                    print(f"\nNumerische Merkmale in Trainings-, Validierungs- und explizitem Testset skaliert (StandardScaler). ({len(cols_to_scale)} Spalten)")
                    # Speichere den Scaler, falls du ihn später für neue Daten brauchst (z.B. für Vorhersagen)
                    # import joblib
                    # joblib.dump(scaler, os.path.join(PROCESSED_DATA_DIR, 'feature_scaler.pkl'))
                    # print("Scaler wurde gespeichert.")
                else:
                    print("\nKeine Spalten zur Skalierung gefunden oder ausgewählt.")

            else:
                print("Nicht genügend Daten in `remaining_data_df` für eine Aufteilung in Training und Validierung.")
                # Setze X_train, y_train auf die vollen Daten, wenn keine Aufteilung möglich ist
                X_train, y_train = X.copy(), y.copy()
                X_val, y_val = pd.DataFrame(), pd.Series(dtype=y.dtype) # Leere Validierungssets
                print(f"Verwende alle {len(X_train)} verbleibenden Samples als Trainingsdaten.")
                # Skalierung auf X_train anwenden
                if not X_train.empty:
                    cols_to_scale = X_train.select_dtypes(include=np.number).columns.tolist()
                    if cols_to_scale:
                        scaler = StandardScaler()
                        X_train[cols_to_scale] = scaler.fit_transform(X_train[cols_to_scale])
                        if not X_test_explicit.empty and all(col in X_test_explicit.columns for col in cols_to_scale):
                            X_test_explicit[cols_to_scale] = scaler.transform(X_test_explicit[cols_to_scale])
                        print(f"\nNumerische Merkmale im Trainingsset skaliert. ({len(cols_to_scale)} Spalten)")


# Überprüfe die Shapes der finalen Datensets
print("\n--- Finale Datenset-Shapes ---")
if 'X_train' in locals(): print(f"X_train: {X_train.shape if not X_train.empty else 'Leer'}")
if 'y_train' in locals(): print(f"y_train: {y_train.shape if not y_train.empty else 'Leer'}")
if 'X_val' in locals(): print(f"X_val: {X_val.shape if not X_val.empty else 'Leer'}")
if 'y_val' in locals(): print(f"y_val: {y_val.shape if not y_val.empty else 'Leer'}")
if 'X_test_explicit' in locals(): print(f"X_test_explicit: {X_test_explicit.shape if not X_test_explicit.empty else 'Leer'}")
if 'y_test_explicit' in locals(): print(f"y_test_explicit: {y_test_explicit.shape if not y_test_explicit.empty else 'Leer'}")
```

```python
# Code Zelle: Phase IV, Schritt 1 - Datenvorbereitung für Sequenzmodelle

import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib
import os

# Annahmen:
# - raw_data_df: Der ursprüngliche, große DataFrame mit den Rohdaten ist verfügbar.
# - VIBRATION_SENSOR_COLUMNS = ['acc_x', 'acc_y', 'acc_z']
# - Die Aufteilung in `remaining_data_df` und `test_set_explicit_df` wurde bereits durchgeführt.
# - PROCESSED_DATA_DIR und RANDOM_SEED sind definiert.

# Stelle sicher, dass die Basis-DataFrames vorhanden sind
if 'remaining_data_df' not in locals() or 'test_set_explicit_df' not in locals():
    print("FEHLER: `remaining_data_df` oder `test_set_explicit_df` nicht gefunden. Führe die vorherigen Zellen zur Datenaufteilung aus.")
    # raise NameError("Benötigte DataFrames für die Fensterung nicht gefunden.")
else:
    print("\n--- Phase IV, Schritt 1: Datenvorbereitung für Sequenzmodelle (Fensterung) ---")

    # --- 1. Konfiguration für die Fensterung ---
    # Diese Werte sind wichtige Hyperparameter!
    WINDOW_SIZE = 4096  # Länge eines Fensters in Samples. 4096 ist eine gute 2er-Potenz.
    STEP_SIZE = WINDOW_SIZE // 2  # Überlappung von 50%. Kleinere Schritte -> mehr Daten, aber höhere Korrelation.

    print(f"Fenstergröße: {WINDOW_SIZE} Samples, Schrittgröße: {STEP_SIZE} Samples (50% Überlappung)")


    # --- 2. Normalisierung der Rohdaten ---
    # Wir passen einen Scaler NUR auf die Trainingsdaten an, um Datenlecks zu vermeiden.
    # Wir nehmen den Teil von remaining_data_df, der für das Training vorgesehen ist.
    
    # Temporärer Split, um den Trainingsanteil zu identifizieren
    # Dies ist nur zur Anpassung des Scalers.
    temp_train_df, _ = train_test_split(remaining_data_df, test_size=0.20, random_state=RANDOM_SEED)

    print(f"\nPasse den Scaler für die Rohdaten auf {len(temp_train_df)} Trainingsdateien an...")
    sequence_scaler = StandardScaler()
    sequence_scaler.fit(temp_train_df[VIBRATION_SENSOR_COLUMNS])

    # Speichere diesen Scaler, er ist entscheidend für die Reproduzierbarkeit
    sequence_scaler_path = os.path.join(PROCESSED_DATA_DIR, 'sequence_raw_data_scaler.pkl')
    joblib.dump(sequence_scaler, sequence_scaler_path)
    print(f"Scaler für Rohdaten-Sequenzen wurde gespeichert: {sequence_scaler_path}")
    
    # Gib den temporären DataFrame frei
    del temp_train_df, _


    # --- 3. Generator-Funktion für das dynamische Erstellen von Batches ---
    def data_generator(df_source: pd.DataFrame, window_size: int, step_size: int, batch_size: int, scaler: StandardScaler):
        """
        Ein Python-Generator, der dynamisch Batches von gefensterten Zeitreihendaten und Labels erstellt.
        
        Args:
            df_source (pd.DataFrame): Der Quell-DataFrame (z.B. Trainings- oder Validierungs-Dateien).
            window_size (int): Die Länge jedes Fensters.
            step_size (int): Die Schrittgröße für die überlappenden Fenster.
            batch_size (int): Die Anzahl der Fenster in einem Batch.
            scaler (StandardScaler): Der bereits angepasste Scaler zur Normalisierung der Rohdaten.
        """
        # Liste aller einzigartigen Dateien im Quell-DataFrame
        files = df_source['filename'].unique()
        
        # Mische die Dateien zu Beginn jeder Epoche für besseres Training
        np.random.shuffle(files)
        
        batch_X, batch_y = [], []
        
        while True: # Endlosschleife, die das Training über mehrere Epochen ermöglicht
            for filename in files:
                # Hole alle Daten für eine Datei
                file_data = df_source[df_source['filename'] == filename]
                
                # Normalisiere die Vibrationsdaten
                # WICHTIG: Erstelle eine Kopie, um die "SettingWithCopyWarning" zu vermeiden
                normalized_vibrations = scaler.transform(file_data[VIBRATION_SENSOR_COLUMNS])
                
                # Das Label ist für alle Fenster dieser Datei gleich
                label = file_data['pitting_level'].iloc[0]

                # Erstelle Fenster aus dieser Datei
                for i in range(0, len(normalized_vibrations) - window_size + 1, step_size):
                    window = normalized_vibrations[i : i + window_size]
                    
                    batch_X.append(window)
                    batch_y.append(label)
                    
                    # Wenn der Batch voll ist, gib ihn zurück (yield)
                    if len(batch_X) == batch_size:
                        yield (np.array(batch_X), np.array(batch_y))
                        # Setze den Batch zurück für die nächste Runde
                        batch_X, batch_y = [], []
            
            # Hinweis: Der Generator stoppt hier, wenn alle Dateien durchlaufen sind.
            # Für das Training mit Keras/TensorFlow ist es üblich, dass der Generator
            # endlos läuft und Keras sich um die Epochengrenzen kümmert.
            # Die `while True`-Schleife stellt dies sicher. Nach einer vollen Runde durch alle
            # Dateien wird die Dateiliste neu gemischt und der Prozess beginnt von vorn.

    
    # --- 4. Erstelle die finalen Trainings-, Validierungs- und Test-DataFrames für den Generator ---
    # Die Aufteilung der *Dateien* haben wir im vorherigen Schritt implizit gemacht.
    # Wir müssen sie hier nur formal in Variablen speichern.
    
    # Trenne die verbleibenden Daten in Trainings- und Validierungs-Dateien
    train_files_df, val_files_df = train_test_split(
        remaining_data_df,
        test_size=0.20, # Gleicher Split-Anteil wie zuvor
        random_state=RANDOM_SEED
    )

    # Das explizite Testset besteht aus den zuvor herausgefilterten Dateien
    test_files_df = test_set_explicit_df

    print("\nAufteilung der Dateien für die Generatoren abgeschlossen:")
    print(f"  Anzahl Trainings-Dateien: {len(train_files_df['filename'].unique())}")
    print(f"  Anzahl Validierungs-Dateien: {len(val_files_df['filename'].unique())}")
    print(f"  Anzahl Test-Dateien: {len(test_files_df['filename'].unique())}")
    
    # --- 5. Teste den Generator ---
    # Es ist eine gute Praxis, den Generator zu testen, um sicherzustellen, dass er
    # Daten mit den korrekten Shapes und Typen liefert.
    
    print("\nTeste den Daten-Generator...")
    BATCH_SIZE = 32 # Typische Batch-Größe
    
    # Erstelle eine Instanz des Generators für die Trainingsdaten
    train_gen_instance = data_generator(
        df_source=train_files_df,
        window_size=WINDOW_SIZE,
        step_size=STEP_SIZE,
        batch_size=BATCH_SIZE,
        scaler=sequence_scaler
    )
    
    # Hole einen einzelnen Batch zum Testen
    try:
        X_batch_sample, y_batch_sample = next(train_gen_instance)
        
        print("Generator-Test erfolgreich!")
        print(f"  Shape von X_batch_sample: {X_batch_sample.shape}")
        print(f"     -> (Batch Größe, Fensterlänge, Anzahl Sensoren)")
        print(f"  Shape von y_batch_sample: {y_batch_sample.shape}")
        print(f"     -> (Batch Größe,)")
        print(f"  Datentyp von X_batch_sample: {X_batch_sample.dtype}")
        print(f"  Datentyp von y_batch_sample: {y_batch_sample.dtype}")
        
    except StopIteration:
        print("FEHLER beim Testen des Generators: Nicht genügend Daten im `train_files_df`, um einen einzigen Batch zu erstellen.")
    except Exception as e:
        print(f"Ein unerwarteter Fehler ist beim Testen des Generators aufgetreten: {e}")
        
# Beende die Zelle, falls die Basis-DataFrames nicht existierten
if 'remaining_data_df' not in locals() or 'test_set_explicit_df' not in locals():
    print("\nAusführung gestoppt, da Basis-DataFrames fehlen.")
```

Merkmale: Es wurde ein umfassender Satz von 42 Merkmalen pro Messdatei extrahiert, darunter:
Zeitbereichsmerkmale: Statistische Momente wie RMS, Kurtosis, Schiefe, Crest-Faktor etc., die die Energie und Impulsivität des Signals beschreiben.
Frequenzbereichsmerkmale: Merkmale aus dem FFT-Spektrum, wie der spektrale Mittelwert (Centroid) und die relative Energie im Frequenzband um die Zahneingriffsfrequenz (GMF).
Automatisierung: Die gesamte Extraktionslogik wurde in wiederverwendbare Funktionen in einer separaten feature_extractor_utils.py-Datei ausgelagert. Dies verbessert die Code-Struktur und löst Kompatibilitätsprobleme mit der Parallelisierung in Jupyter Notebooks.
Performance-Optimierung: Der Extraktionsprozess für die 390 Dateien wurde mithilfe des concurrent.futures.ProcessPoolExecutor-Moduls parallelisiert. Dies verteilt die rechenintensive Arbeit auf mehrere CPU-Kerne und reduziert die Verarbeitungszeit drastisch.
4.4. Datenaufbereitung für die Modellierung
Der resultierende Merkmals-DataFrame (features_df) wurde sorgfältig für die Modellierung vorbereitet:

NaN-Behandlung: Eventuell entstandene NaN-Werte (z.B. durch Division durch Null) wurden mittels Imputation (Median-Strategie) behandelt.
Aufteilung in Trainings-, Validierungs- und Testsets:
Explizites Testset: Gemäß den Vorgaben des Wettbewerbs wurden die Dateien mit pitting_level 5, 7, 9, 10 als separates, ungesehenes Testset zurückgehalten.
Train/Val-Split: Die verbleibenden Daten wurden in ein Trainingsset (80%) und ein Validierungsset (20%) aufgeteilt.
Merkmalskalierung: Die numerischen Merkmale wurden mit StandardScaler skaliert. Der Scaler wurde nur auf den Trainingsdaten angepasst (fit) und dann konsistent auf Validierungs- und Testdaten angewendet (transform), um Datenlecks zu verhindern.
4.5. Merkmalsauswahl (Feature Selection)
Um die Dimensionalität zu reduzieren und die relevantesten Prädiktoren zu identifizieren, wurde ein systematischer Auswahlprozess durchgeführt:

VarianceThreshold: Entfernung von Merkmalen mit sehr geringer Varianz.
SelectKBest: Univariate Auswahl der Top-Merkmale basierend auf ihrer statistischen Beziehung zur Zielvariablen (pitting_level).
SelectFromModel mit RandomForestRegressor: Eine Embedded-Methode, die die inhärente Merkmalswichtigkeit eines Random Forest nutzt, um die finale, leistungsstarke Untermenge an Merkmalen auszuwählen.
4.6. Geplante Fortgeschrittene Modellierung und XAI
Nachdem die Daten aufbereitet und eine Baseline etabliert wurde, folgt der Kern des Projekts:

Datenvorbereitung für Sequenzmodelle: Die Rohdaten werden mittels einer Generator-Funktion dynamisch in überlappende Fenster (Sequenzen) zerlegt. Dieser speichereffiziente Ansatz liefert die Eingabedaten für die folgenden Modelle.
Transformer-Modell: Implementierung einer Transformer-Architektur, die direkt auf den Zeitreihenfenstern operiert, um zeitliche Abhängigkeiten zu lernen und das pitting_level vorherzusagen.
Physics-Informed Neural Network (PINN): Entwicklung eines neuronalen Netzes, dessen Verlustfunktion durch Terme ergänzt wird, die auf physikalischen Prinzipien des Getriebes basieren (z.B. Korrelation von Vibrationsenergie mit Last und Drehzahl).
Explainable AI (XAI):
Analyse des besten Baseline-Modells mit SHAP, um die Wichtigkeit der extrahierten Merkmale zu verstehen.
Analyse des Transformers durch Visualisierung der Attention Maps, um zu sehen, auf welche Teile der Schwingungssequenz das Modell seine "Aufmerksamkeit" richtet.
